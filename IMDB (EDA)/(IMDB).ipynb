# Classify Movies Based on Ratings [Excellent, Good, and Average]


```python
def rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating>=6.0:
        return "Good"
    else:
        return "Average"
    
```

```python
df['rating_cat']=df['Rating'].apply(rating)
```

# Find Unique Values From Genre 


```python
df.columns
```

```python
df['Genre']
```

```python
list1=[]
for value in df['Genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
one1=[]
for item in list1:
    for item1 in item:
        one1.append(item1)
```

```python
uniq1=[]
for item in one1:
    if item not in uniq1:
        uniq1.append(item)
```

```python
uniq1
```
